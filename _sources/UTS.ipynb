## UTS Naive Bayes

Naïve Bayes Classifier merupakan sebuah metoda klasifikasi yang berakar pada teorema Bayes . Metode pengklasifikasian dg menggunakan metode probabilitas dan statistik

```python
# Naive Bayes Classification

# Importing the libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
```

```python
data = "https://raw.githubusercontent.com/kb22/Heart-Disease-Prediction/master/dataset.csv"
datauts = pd.read_csv(data)
```

```python
#looking at the first 5 values of the datauts
datauts.head()
```

```python
#periksa jumlah baris dan kolom dalam dataset
datauts.shape
```

```python
#Pilih data menjadi variabel independen 'X' dan variabel 'y'
X = data.drop('target', axis=1)
y = data['target']
# Memisahkan datauts ke dalam set Pelatihan dan set Testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# lakukan pelatihan model dengan menggunakan Naive Bayes Classifier:
nb = GaussianNB()
nb.fit(X_train, y_train)
```

```python
# Setelah model terlatih, kita dapat melakukan prediksi pada data testing dan menghitung akurasi model:
y_pred = nb.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy:', accuracy)
```

## UTS K-NN

Algoritma K-Nearest Neighbor (KNN) merupakan algoritma machine learning sederhana dan mudah diterapkan yang dapat digunakan untuk menyelesaikan masalah klasifikasi dan regresi. Algoritma ini termasuk dalam jenis supervised learning.

```python
# Klasifikasi using K-NN

# Importing the libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
# Selanjutnya, kita baca dataset dan bagi data menjadi data training dan data testing:
data = pd.read_csv('https://raw.githubusercontent.com/kb22/Heart-Disease-Prediction/master/dataset.csv')
X = data.drop('target', axis=1)
y = data['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Kemudian, kita lakukan pelatihan model dengan menggunakan K-NN:
k = 5
knn = KNeighborsClassifier(n_neighbors=k)
knn.fit(X_train, y_train)
```

```python
# Setelah model terlatih, kita dapat melakukan prediksi pada data testing dan menghitung akurasi model:
y_pred = knn.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy:', accuracy)
```

```python
# Dalam klasifikasi dengan K-NN, nilai k yang tepat dapat mempengaruhi akurasi model. Oleh karena itu, kita dapat mencoba beberapa nilai k yang berbeda dan memilih nilai k yang memberikan akurasi tertinggi:

for k in range(1, 21):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print('k =', k, ':', 'Accuracy:', accuracy)

# Dari hasil percobaan di atas, terlihat bahwa akurasi tertinggi diperoleh pada k=7 dengan akurasi sekitar 70%.

# Dari hasil analisis di atas, dapat disimpulkan bahwa Naive Bayes Classifier memberikan akurasi yang lebih baik daripada K-NN untuk dataset ini. Namun, nilai k yang tepat pada K-NN dapat meningkatkan akurasi model.
```
